```python
# Ensure product_name, brand, and product_category_tree are strings
df['product_name'] = df['product_name'].astype(str)
df['brand'] = df['brand'].astype(str)
df['product_category_tree'] = df['product_category_tree'].astype(str)

# Function to check if the brand name is present in product_name or product_category_tree
def check_brand_in_product_or_category(product_name, brand_name, category_tree):
    return (brand_name.lower() in product_name.lower()) or (brand_name.lower() in category_tree.lower())

# Apply the function
df['is_brand_present'] = df.apply(
    lambda row: check_brand_in_product_or_category(row['product_name'], row['brand'], row['product_category_tree']),
    axis=1
)

# Check the result
print(df[['product_name', 'brand', 'product_category_tree', 'is_brand_present']].head())

# Count the number of matches
print("Number of products where the brand name is present in product name or product category tree:", df['is_brand_present'].sum())
```

```python
import pandas as pd
import ast
import json

df['product_specifications'] = df['product_specifications'].astype(str)


# Function to clean and parse the specifications
def clean_and_parse_specs(specs_str):
    # Replace non-standard characters and adjust the string
    specs_str = specs_str.replace('=>', ':').replace('“', '"').replace('”', '"')
    try:
        # Convert string to dictionary
        specs_dict = ast.literal_eval(specs_str)
        if isinstance(specs_dict, dict) and 'product_specification' in specs_dict:
            return {item['key']: item['value'] for item in specs_dict['product_specification'] if 'key' in item and 'value' in item}
        return {}
    except (SyntaxError, ValueError):
        return {}

# Apply the function to clean and parse the specifications
df['extracted_specifications'] = df['product_specifications'].apply(clean_and_parse_specs)


```

```python

import numpy as np
# Replace brand names in cleaned_list with NaN

df['brand'] = df['brand'].apply(lambda x: np.nan if x in cleaned_list else x)

# Drop the specified columns
y_df = df.drop(columns=['uniq_id', 'pid', 'overall_rating', 'is_brand_present', 'product_specifications','is_rating_same'])

# Display the updated DataFrame
y_df.head()
```

```python
import pandas as pd
import re

# Function to extract price from the description
def extract_price(description):
    if isinstance(description, str):
        match = re.search(r'Rs\. (\d+)', description)
        if match:
            return int(match.group(1))
    return None

# Fill missing 'discounted_price' and 'retail_price' using the extracted price from 'description'
y_df['extracted_price'] = y_df['description'].apply(extract_price)

# Update 'discounted_price' and 'retail_price' where they are null
y_df['discounted_price'].fillna(y_df['extracted_price'], inplace=True)
y_df['retail_price'].fillna(y_df['extracted_price'], inplace=True)

# Drop the temporary 'extracted_price' column
y_df.drop(columns=['extracted_price'], inplace=True)


```
